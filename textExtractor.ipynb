```
import os
import cv2

if not os.path.exists('fortun'):
    os.makedirs('fortun')

test_vid = cv2.VideoCapture('Task_4_Moving_pipe_OCR.mp4')

index = 0
while test_vid.isOpened():
    ret,frame = test_vid.read()
    if not ret:
        break


    name = './fortun/frame' + str(index) + '.png'

    print ('Extracting frames...' + name)
    cv2.imwrite(name, frame)
    index = index + 1
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
    
test_vid.release()
```
